## Step 2: Build Neural Network

To build the neural network, you can refer to your own codes you wrote while following the [Deep Learning with Python, TensorFlow, and Keras tutorial](https://www.youtube.com/watch?v=wQ8BIBpya2k) in the lesson. It's pretty similar to what you will be doing in this lab.

1. Split the training and test data.
2. Create a `Sequential` model.
3. Add several layers to your model. Make sure you use ReLU as the activation function for the middle layers. Use Softmax for the output layer because each output has a single label and all the label probabilities add up to 1.
4. Compile the model using `adam` as the optimizer and `sparse_categorical_crossentropy` as the loss function. For metrics, use `accuracy` for now.
5. Fit the training data.
6. Evaluate your neural network model with the test data.
7. Save your model as `tic-tac-toe.model`.

````python
import tensorflow as tf
print(tf.__version__)
print(tf.config.list_physical_devices())
````

````python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from keras.optimizers import Adam
````

````python
from sklearn.model_selection import train_test_split as tts
````

### Task 1

````python
# your code here
# TASK 1 spliting data
X_train, X_test, y_train, y_test= tts(X_scaled, y, test_size= .2, random_state= 69)

print(f"X-shape: {X.shape} --- X_train-shape: {X_train.shape} --- X_test-shape: {X_test.shape}")
print(f"y-shape: {y.shape} --- y_train-shape: {y_train.shape} --- y_test-shape: {y_test.shape}")
````

### Task 2

````python
# TASK 2 creating sequential model
model= Sequential()
````

````python
# EXTRAS for tuning

from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.layers import Dropout
````

### Task 3

````python
# TASK 3 adding several layers

# input layer (h1)
model.add(Dense(units= 64, activation= "relu", input_shape= (27,)))
model.add(Dropout(.1))

# middle layer (h2)
model.add(Dense(units= 32, activation= "relu"))
model.add(Dropout(.1))

# output layer (h3)
model.add(Dense(units= 2, activation= "softmax"))
````

### Task 4

````python
# TASK 4 compiling the model
model.compile(optimizer= "adam",  # Adam as optimizer
              loss= "sparse_categorical_crossentropy", # Cat. Crossentropy as loss function
              metrics= ["accuracy"] # Accuracy as metrics
)
````

````python
early_stop= EarlyStopping(monitor="val_loss", patience=6, restore_best_weights=True)
````

### Task 5

````python
# TASK 5 fit model
model.fit(X_train, y_train, 
          epochs= 80,
          batch_size= 8,
          validation_split= .1,
          callbacks= [early_stop],
          verbose= True
)
````

### Task 6

````python
# TASK 6 evaluate model

test_loss, test_acc= model.evaluate(X_test, y_test)
print(test_acc)
print(test_loss)
````

### Task 7

````python
# TASK 7 save the model
# When trying to save as .model raise the error:
# ValueError: Invalid filepath extension for saving. Please add either a `.keras` extension for the native Keras format (recommended) or a `.h5` 
# extension. Use `model.export(filepath)` if you want to export a SavedModel for use with TFLite/TFServing/etc. Received: filepath=tic-tac-toe.model.

model.save("tic-tac-toe.keras")
````
